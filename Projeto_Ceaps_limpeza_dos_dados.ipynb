```python
#importando a biblioteca utilizada para limpeza de dados
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importando os dados de 2022
data_2022 = pd.read_csv('despesa_ceaps_2022.csv', encoding= "cp1252", sep= ';')
data_2022.head()
```

```python
#importando os dados de 2021
data_2021= pd.read_csv('despesa_ceaps_2021.csv', encoding= "cp1252", sep= ';')
data_2021.head(3)
```

```python
#importando os dados de 2020
data_2020= pd.read_csv('despesa_ceaps_2020.csv', encoding= "cp1252", sep= ';')
data_2020.head(3)
```

```python
#importando os dados de 2019
data_2019= pd.read_csv('despesa_ceaps_2019.csv', encoding= "cp1252", sep= ';')
data_2019.head(3)
```

Utilizando a fun√ß√£o concat para  unir os quatro datasets em apenas um para facilitar mais a an√°lise

```python
#concatenando os dados de 2019,2020, 2021 e 2022 em um dataset 
data_concatenado= pd.concat([data_2022, data_2021, data_2020, data_2019], axis=0)
data_concatenado
```

Podemos percorrer o dataset e observar os valores √∫nicos que ocorrem em cada coluna 

```python
for var in data_concatenado:
  print('Coluna', var, 'Tipo da coluna', data_concatenado[var].dtype)
  print('Valores √∫nicos', data_concatenado[var].unique(), '\n')
```

Vamos verificar se existem colunas duplicadas 

```python
data_concatenado.duplicated()
```

As colunas: 'C√ìD_DOCUMENTO', 'DOCUMENTO' e 'DETALHAMENTO' s√£o irrelevantes na an√°lise que ser√° feita por isso elas ser√£o deletadas com o m√©todo drop. Por padr√£o o m√©todo drop apaga as linhas no nosso caso ser√£o apagadas as colunas para isso basta especificar axis =*1*


```python
data_concatenado = data_concatenado.drop(['COD_DOCUMENTO', 'DOCUMENTO', 'DETALHAMENTO'], axis =1)
data_concatenado.head(3)
```

```python
data_concatenado
```

Verificando se existem dados do tipo Not a Number(NaN)

```python
data_concatenado.isna().sum()
```

```python
data_concatenado.isnull().sum()
```

Verificando os tipos de vari√°veis que foi poss√≠vel identificar

```python
#tipos de vari√°veis de cada coluna
data_concatenado.info()
```

Podemos perceber que algumas  vari√°veis est√£o sendo lidas de maneira incorreta para isso vamos utilizar o m√©todo astype para atribuir os valores corretos as colunas

```python
#Senador, tipo de despesa, fornecedor s√£o strings
data_concatenado["SENADOR"] = data_concatenado["SENADOR"].astype('string')
data_concatenado['TIPO_DESPESA']= data_concatenado['TIPO_DESPESA'].astype('string')
data_concatenado['FORNECEDOR']= data_concatenado['FORNECEDOR'].astype('string')

#Valor reembolsado √© do tipo float para isso precisamos transforma as v√≠rgulas em pontos e permanecer os pontos em pontos
#Na coluna reembolso precisamos utilizar uma fun√ß√£o para assim transformar a coluna no tipo certo que seria float
data_concatenado['VALOR_REEMBOLSADO'] = data_concatenado['VALOR_REEMBOLSADO'] .apply(lambda x: float(x.replace(".","").replace(",",".")))
```
